```python
from PyPDF2 import PdfReader
import pandas as pd
import re
```

```python
resume_path = "resume.pdf"
```

```python
reader = PdfReader("resume.pdf")

text = ""

for page in reader.pages:
    text += page.extract_text()

print(text[:2000])
```

```python
email = re.findall(
    r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
    text
)

phone = re.findall(
    r"\d{11}",
    text
)

print("Email:", email)
print("Phone:", phone)
```

```python
skills_db = [
    "Python",
    "C++",
    "Java",
    "SQL",
    "HTML",
    "CSS",
    "JavaScript",
    "Machine Learning",
    "Deep Learning",
    "Power BI",
    "Git",
    "GitHub"
]
```

```python
found_skills = []

for skill in skills_db:
    if skill.lower() in text.lower():
        found_skills.append(skill)

print("Skills Found:")
print(found_skills)
```

```python
score = 0

if email:
    score += 20

if phone:
    score += 20

score += len(found_skills) * 5

if score > 100:
    score = 100

print("Resume Score =", score)
```

```python
missing_skills = []

for skill in skills_db:
    if skill not in found_skills:
        missing_skills.append(skill)

print("Missing Skills:")
print(missing_skills)
```

```python
suggestions = []

if len(found_skills) < 5:
    suggestions.append(
        "Add more technical skills."
    )

if not email:
    suggestions.append(
        "Add email address."
    )

if not phone:
    suggestions.append(
        "Add phone number."
    )

suggestions.append(
    "Add projects section."
)

suggestions.append(
    "Add certifications."
)

print("Suggestions:")
for s in suggestions:
    print("-", s)
```

```python
print("="*50)
print("AI RESUME ANALYZER REPORT")
print("="*50)

print("\nResume Score:", score, "/100")

print("\nEmail:")
print(email)

print("\nPhone:")
print(phone)

print("\nSkills Found:")
for skill in found_skills:
    print("✓", skill)

print("\nMissing Skills:")
for skill in missing_skills:
    print("✗", skill)

print("\nSuggestions:")
for s in suggestions:
    print("•", s)
```
